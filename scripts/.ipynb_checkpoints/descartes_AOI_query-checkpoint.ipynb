## most recent for pulling rasters from descartes

```python
import os,sys
import numpy as np
import descarteslabs as dl
import pyproj
import matplotlib.pyplot as plt
sys.path.append('/home/kumarv/khand035/.conda/envs/gdalenv/lib/python2.7/site-packages/')
from osgeo import gdal,ogr
raster_client = dl.Raster()
```

```python
# Create an aoi feature to clip imagery to, 

#longitude first, latitude second 
#polygon drawing: upper left--> upper right--> lower right--> lower left --> back to upper left
aoi_geometry = {'type': 'Polygon',
                'coordinates': (((27.5151, 9.0295),
                                 (27.5849, 9.0295),
                                 (27.5849, 9.0666),
                                 (27.5151, 9.0666),
                                 (27.5151, 9.0295)),)}
```

```python
def search_AOI(box):
#date time must have standard format: 2018-07-01, NOT 2018-7-1
    scenes, geoctx = dl.scenes.search(box,
                               products=["sentinel-1:GRD"],
                               #products=['airbus:oneatlas:phr:v2'],
                               start_datetime="2017-01-01",
                               end_datetime="2017-01-31")

    return scenes, geoctx
```

```python
scenes, geoctx=search_AOI(aoi_geometry)
print('Number of scenes found: {}'.format(len(scenes)))
```

```python
# Pick just one scene, start with most recent one
scene=scenes[2]
arr = scene.ndarray("vv vh vv", geoctx,mask_alpha=False)
# Display the scene
dl.scenes.display(arr, size=4, title=scene.properties.id)
arr.shape
```

# The Raster ndarray API call to download an image (if multiple images are specified then mosaic can be performed). Other parameters are available

```python
select_scene=scene.properties.id
s2_arr_wgs84, m_wgs84 = raster_client.ndarray(select_scene,
                                             srs="EPSG:4326",  # WGS84 (lat-lon)
                                              # 0.0000046 corresponds to 50cm in decimal degrees,
                                              #0.0001 corresponds to 10m
                                             resolution=0.0001, 
                                             bands=["red", "green", "blue"],
                                             cutline=aoi_geometry,
                                             data_type="UInt16")
```

```python
arr2=np.rollaxis(s2_arr_wgs84, 2, 0)
dl.scenes.display(arr2, size=8, title=select_scene)
arr2.shape
```

# save as EPSG:4326

```python
dst_filename='descartes/bahr_2018-09-26.tif'
x_pixels = s2_arr_wgs84.shape[1]  # number of pixels in x
y_pixels = s2_arr_wgs84.shape[0]   # number of pixels in y
driver = gdal.GetDriverByName('GTiff')
dst_ds = driver.Create(dst_filename,x_pixels, y_pixels, 3,gdal.GDT_UInt16) #GDT_Byte is Uint8

r_pixels=s2_arr_wgs84[:,:,0]
g_pixels=s2_arr_wgs84[:,:,1]
b_pixels=s2_arr_wgs84[:,:,2]

geotransform=m_wgs84['geoTransform']
dst_ds.SetGeoTransform(geotransform)
# specify coords
srs = osr.SpatialReference()            # establish encoding
srs.ImportFromEPSG(4326)                # WGS84 lat/long
dst_ds.SetProjection(srs.ExportToWkt()) # export coords to file
dst_ds.GetRasterBand(1).WriteArray(r_pixels)   # write r-band to the raster
dst_ds.GetRasterBand(2).WriteArray(g_pixels)   # write g-band to the raster
dst_ds.GetRasterBand(3).WriteArray(b_pixels)   # write b-band to the raster
dst_ds.FlushCache()                     # write to disk
dst_ds = None
```
